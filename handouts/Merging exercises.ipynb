# The merge function


```python
import pandas as pd

left_df = pd.DataFrame({
                    'firm': ['Accenture','Citi','GS'],
                    'varA': ['A1', 'A2', 'A3']})

right_df = pd.DataFrame({
                    'firm': ['GS','Chase','WF'],
                    'varB': ['B1', 'B2', 'B3'],
                    'varc': ['C1', 'C2', 'C3']})    
```

```python
left_df.merge(right_df)
```

The main parameters:
- right
- how
- on and its variants 
- indicator 
- validate

```python
ChosenOne = pd.DataFrame({
                    'tic': ['GS','GS','GS'],
                    'varB': ['B1', 'B2', 'B3'],
                    'varc': ['C1', 'C2', 'C3']})    

CurryForThree = pd.DataFrame({
                    'var1': ['GS','GS','GS'],
                    'varD': ['D1', 'D2', 'D3'],
                    'varE': ['E1', 'E2', 'E3']}).set_index('var1')
```

**Q0**

```python
left_df.merge(right_df,how = 'inner') 
# inner: the default only gives back the data with keys in both
# the "key" in a merge is the variables you merge on
# what is the key here? FIRM and .merge() guessed based on overlapping column names
```

```python
left_df.merge(right_df,how = 'outer')

# keep all keys in both datasets
# "outer" inner + all unmatched obs in left and right
```

```python
left_df.merge(right_df,how = 'left')
```

```python
left_df.merge(right_df,how = 'right')
```

```python
left_df.merge(right_df,how = 'cross')
```

```python
left_df.merge(CurryForThree, how= 'outer', left_on= 'firm', right_on= 'var1')

left_df.merge(CurryForThree, how= 'outer', left_on= 'firm', right_index= True)
```

**Q3**

```python
left_df.merge(right_df, how= 'outer', indicator= True)

# you can use indicate to get a variable that tells you the source on an object


# heres how it looks
a_bad_name = left_df.merge(right_df, how= 'outer', indicator= True)
a_bad_name
```

**Q4**

```python
left_df.merge(right_df, how= 'outer', indicator= True, validate= '1:1')

# validate is going to help check that the merge you are doing is what to want. 3.4.1.1.
# do not use many to many
```

```python
left_df.merge(right_df, how= 'outer', indicator= True, validate= '1:m')
```

```python
left_df.merge(right_df, how= 'outer', indicator= True, validate= 'm:1')
```

```python
left_df.merge(right_df, how= 'outer', indicator= True, validate= 'm:m')
```
